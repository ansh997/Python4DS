```python
# select Fertility rate, total (births per woman) for the India
hist_indicator = 'Fertility rate, total \(births'
hist_country = 'IND'
hist_indicator1 = 'Mortality rate, infant \(per'

mask1 = data['IndicatorName'].str.contains(hist_indicator)
mask3 = data['IndicatorName'].str.contains(hist_indicator1)
mask2 = data['CountryCode'].str.contains(hist_country)

# stage is just those indicators matching the IND for country code
stage = data[mask1 & mask2]
stage1 = data[mask2 & mask3]
```

```python
stage.head()
```

```python
# get the years
years = stage['Year'].values
# get the values 
fert_rate = stage['Value'].values

# create
plt.bar(years,fert_rate)
plt.show()
```

Birth rate per woman has decreased significantly because of increase in litteracy rate and decrease in infant mortality rate.

```python
# get the years
years1 = stage1['Year'].values
#mort_rate
mort_rate = stage1['Value'].values

plt.bar(years1, mort_rate)
plt.show()
```
